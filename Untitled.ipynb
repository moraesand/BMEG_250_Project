## Section I: Define the Biological Problem

**The methionine / SAM / SAH cycle and AHCY**

The methionine cycle is a central metabolic pathway responsible for maintaining the cellular supply of methyl groups. The amino acid methionine is first adenylated by methionine adenosyltransferase (MAT) to form S-adenosylmethionine (SAM), the primary methyl-group donor for over 200 methyltransferase reactions in human cells — including methylation of DNA, RNA, histones, and small molecules. Each time a methyltransferase transfers a methyl group from SAM to a substrate, SAM is converted to S-adenosylhomocysteine (SAH), a potent competitive inhibitor of SAM-dependent methyltransferases that occupies the same active-site pocket as SAM but cannot donate a methyl group. S-adenosylhomocysteine hydrolase (AHCY/SAHH) is the sole mammalian enzyme capable of relieving this inhibition — it catalyzes the hydrolysis of SAH into adenosine and homocysteine using NAD⁺ as a cofactor, thereby regenerating the substrate pool and sustaining the SAM/SAH ratio that governs the cell's overall methylation potential. The homocysteine released by AHCY is either remethylated back to methionine (completing the cycle) or diverted into the transsulfuration pathway; because no other enzyme can remove SAH, AHCY activity is absolutely indispensable for all SAM-dependent methylation in the cell.

**Why inhibiting AHCY is an antiviral strategy for MERS-CoV**

Like all coronaviruses, MERS-CoV must produce a properly methylated 5′ RNA cap on every viral mRNA in order to facilitate translation by host ribosomes and to mimic host mRNA structure, thereby evading innate immune sensors (e.g., RIG-I, MDA5) that detect uncapped or aberrantly capped RNA. MERS-CoV achieves cap methylation through two SAM-dependent nonstructural proteins — nsp14 (N7-methyltransferase) and the nsp10/nsp16 complex (2′-O-methyltransferase) — both of which consume SAM and release SAH as a byproduct. Because the virus encodes no AHCY of its own, it depends entirely on host AHCY to continuously clear this SAH and replenish SAM; pharmacological inhibition of human AHCY therefore causes intracellular SAH to accumulate and competitively inhibit the viral methyltransferases the virus requires for capping. The resulting impairment of viral RNA cap methylation reduces translational efficiency of viral transcripts and renders viral RNA recognizable by pattern-recognition receptors, both of which suppress replication. This host-directed mechanism is particularly attractive because it targets a conserved host enzyme rather than a rapidly mutating viral protein, raising the barrier to resistance.

The AlphaFold2 model of human AHCY was generated using ColabFold and produced five ranked predictions, all of which show near-identical confidence profiles (pLDDT per position plot), indicating a highly consistent and confident prediction. The predicted IDDT plot shows pLDDT scores consistently above 90 across the majority of the 432-residue sequence, with only a few short loop regions dipping to around 80 (confident) and a very small N-terminal segment showing lower confidence (~50 or below, shown in red/green in the structure viewer). This overall very high pLDDT indicates AlphaFold is highly certain about the predicted fold, which is expected for a well-conserved housekeeping enzyme like AHCY that has abundant homologous sequences in the training database (as confirmed by the sequence coverage plot showing over 15,000 aligned sequences across most positions).

Despite the high confidence of the AlphaFold model, the experimental crystal structure 1LI4 is preferred as the primary docking receptor for this project. The crystal structure provides an experimentally validated atomic conformation resolved at 2.01 Å, and critically retains the NAD+ cofactor in its native binding geometry adjacent to the active site - a feature absent in the AlphaFold model, which predicts structure from sequence alone without any cofactor or ligand context. The presence of NAD+ in 1LI4 is important for realistic docking because it partially defines the shape and electrostatic environment of the binding pocket where aristeromycin analogues are expected to bind. The AlphaFold model is nonetheless valuable as a complementary resource: it confirms that the overall predicted fold is highly consistent with the crystal structure, validates full-length sequence coverage with no missing terminal residues, and provides per-residue confidence scores that highlight any structurally uncertain regions. Together, these two models support the robustness of the receptor used in this study.

## Section XIV: Convert ΔG to Approximate K_D

The relationship between binding free energy and dissociation constant is:

$$\Delta G = RT \ln K_D \quad \Longrightarrow \quad K_D = e^{\Delta G / RT}$$

where:
- $R = 1.987 \times 10^{-3}$ kcal mol⁻¹ K⁻¹ (gas constant)
- $T = 298$ K (room temperature, ~25 °C)

````python
import numpy as np
import pandas as pd

# ── Fill in your SwissDock ΔG values below ────────────────────────────────────
compounds = {
    "1 (aristeromycin)": None,   # replace None with ΔG in kcal/mol, e.g. -8.5
    "2a":               None,
    "2c":               None,
    "3a":               None,
    # Add more as needed:
    # "3b":             None,
}

# Constants
R = 1.987e-3   # kcal / (mol·K)
T = 298.0      # K

results = []
for name, dG in compounds.items():
    if dG is not None:
        kD_M  = np.exp(dG / (R * T))        # K_D in Molar
        kD_uM = kD_M * 1e6                   # K_D in µM
        kD_nM = kD_M * 1e9                   # K_D in nM
        results.append({
            "Compound": name,
            "ΔG (kcal/mol)": dG,
            "K_D (µM)": round(kD_uM, 4),
            "K_D (nM)": round(kD_nM, 2),
        })
    else:
        results.append({
            "Compound": name,
            "ΔG (kcal/mol)": "TBD",
            "K_D (µM)": "TBD",
            "K_D (nM)": "TBD",
        })

kd_df = pd.DataFrame(results)
print(kd_df.to_string(index=False))
````
